### 8. Create new columns for the year and month of the tender award

```python
gebiz["year"] = gebiz["award_date"].apply(lambda x: x.split("-")[0])
```

```python
gebiz["month"] = gebiz["award_date"].apply(lambda x: x.split("-")[1])
```

```python
gebiz.head()
```

### 9. List the top five months in 2016 which had the most number of tenders awarded to suppliers

```python
gebiz_2016_awarded = gebiz[(gebiz.tender_detail_status == "Awarded to Suppliers") & (gebiz.year == "2016")]
```

```python
gebiz_2016_awarded.groupby("month")["tender_description"].size().sort_values(ascending=False)[:5]
```

### 10. List the top ten agency-supplier pairs by number of tenders awarded

```python
gebiz[gebiz.tender_detail_status == "Awarded to Suppliers"].groupby(["agency", "supplier_name"]).size().sort_values(ascending=False)[:10]
```
